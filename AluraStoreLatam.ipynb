### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres = ["Tienda 1","Tienda 2","Tienda 3","Tienda 4"]
```

#1. Análisis de facturación



```python
ingresos = [t["Precio"].sum() for t in tiendas]

for nombre, ingreso in zip(nombres, ingresos):
    print(nombre, ingreso)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(nombres, ingresos)
plt.title("Ingresos por tienda")
plt.ylabel("Ingresos")
plt.show()
```

# 2. Ventas por categoría

```python
categorias = []

for t in tiendas:
    categorias.append(t.groupby("Categoría del Producto")["Precio"].sum())

df_categorias = pd.DataFrame(categorias, index=nombres)

print(df_categorias)
```

```python
categorias = []

for t in tiendas:
    categorias.append(t.groupby("Categoría del Producto")["Precio"].sum())

df_categorias = pd.DataFrame(categorias, index=nombres)

print(df_categorias)
```

# 3. Calificación promedio de la tienda


```python
calificaciones = [t["Calificación"].mean() for t in tiendas]

for nombre, c in zip(nombres, calificaciones):
    print(nombre, round(c,2))
```

```python
plt.bar(nombres, calificaciones)
plt.title("Calificación promedio por tienda")
plt.ylabel("Calificación")
plt.show()
```

# 5. Envío promedio por tienda

```python
envio = [t["Costo de envío"].mean() for t in tiendas]

for nombre, e in zip(nombres, envio):
    print(nombre, round(e,2))
```

```python
plt.bar(nombres, envio)
plt.title("Costo promedio de envío")
plt.ylabel("Costo de envío")
plt.show()
```

Conclusión

Introducción
El objetivo de este análisis es evaluar el desempeño de las cuatro tiendas de la cadena Alura Store con el fin de identificar cuál presenta el menor rendimiento y, por lo tanto, debería ser vendida por el Sr. Juan para financiar un nuevo emprendimiento.

Para realizar este análisis se utilizaron datos de ventas de cada tienda, considerando diferentes indicadores clave como los ingresos totales generados, las categorías de productos con mayor y menor volumen de ventas, las calificaciones promedio de los clientes, los productos más vendidos y el costo promedio de envío. A partir de estos datos se generaron diversas visualizaciones que permiten comparar el desempeño de las tiendas y facilitar la toma de decisiones basada en datos.
Análisis de ingresos

El análisis de los ingresos totales permite identificar qué tiendas generan mayor volumen de ventas. Los resultados muestran que algunas tiendas tienen un desempeño más sólido en términos de facturación, mientras que otras presentan ingresos considerablemente menores.
A partir de los gráficos generados se observa que la Tienda 4 registra el menor ingreso total entre las cuatro tiendas, lo que sugiere un menor nivel de actividad comercial en comparación con las demás.

Análisis de categorías de productos
El análisis de las categorías de productos permite identificar qué tipos de artículos generan mayores ventas dentro de cada tienda. En general, algunas categorías concentran gran parte de los ingresos, lo que indica una mayor demanda por parte de los clientes.

Sin embargo, también se identificaron categorías con menor volumen de ventas, lo que puede indicar oportunidades de mejora en la estrategia comercial o menor interés del mercado en determinados productos.

Análisis de satisfacción del cliente
Las calificaciones promedio de los clientes constituyen un indicador importante del nivel de satisfacción con la experiencia de compra. En general, las tiendas presentan calificaciones similares, lo que indica que la calidad del servicio es relativamente consistente dentro de la cadena.
No obstante, este indicador por sí solo no compensa las diferencias observadas en el rendimiento de ventas entre las tiendas.

Análisis de productos vendidos
El análisis de los productos más y menos vendidos permite identificar cuáles artículos impulsan el volumen de ventas en cada tienda. Este análisis muestra que ciertos productos tienen una rotación significativamente mayor, contribuyendo de forma importante a los ingresos totales.

Asimismo, algunos productos presentan niveles de ventas más bajos, lo cual podría indicar menor demanda o una menor visibilidad dentro del catálogo.

Análisis del costo de envío
El costo promedio de envío también puede influir en la decisión de compra de los clientes. Costos de envío más elevados pueden representar una barrera para los consumidores, mientras que costos más bajos pueden resultar más atractivos y favorecer las ventas.

En el análisis realizado se observaron diferencias moderadas entre las tiendas en cuanto a este indicador, aunque este factor no parece ser el principal determinante de las diferencias en ingresos.

Conclusión y recomendación

Con base en el análisis de los ingresos, las categorías de productos, las calificaciones de los clientes, los productos vendidos y el costo promedio de envío, se concluye que la Tienda 4 presenta el menor desempeño general dentro de la cadena Alura Store.

En particular, la Tienda 4 registra los ingresos más bajos, lo que indica un menor volumen de ventas en comparación con las otras tiendas analizadas. Aunque las calificaciones de los clientes y los costos de envío no muestran diferencias extremas, el rendimiento económico general de esta tienda es inferior.

Por lo tanto, se recomienda que el Sr. Juan venda la Tienda 4, ya que representa la opción con menor desempeño dentro del conjunto de tiendas evaluadas. Esta decisión permitiría optimizar los recursos disponibles y concentrar los esfuerzos en las tiendas con mejores resultados comerciales.
